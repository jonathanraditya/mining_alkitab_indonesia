```python
import markdown
import json
import pdfkit
```

```python
RESULTS_FILE = './alkitab_data.json'
MARKDOWN_FILE = './alkitab_data_v2.md'
with open(RESULTS_FILE, 'r') as f:
    alkitab_data = json.load(f)
    
markdown_results = ''
for kitab_key in alkitab_data.keys():
    # Add Kitab title
    markdown_results = markdown_results + f'#{kitab_key}\n'
    for bab_key in alkitab_data[kitab_key].keys():
        # Add bab sub-title
        markdown_results = markdown_results + f'##{kitab_key} Bab {int(bab_key)+1}\n'
        ayat_count = 1
        for ayat in alkitab_data[kitab_key][bab_key]:
            markdown_results = markdown_results + f'{ayat_count}. {ayat}\n'
            ayat_count+=1
    
    # Add horizontal lines for separation between bab
    markdown_results = markdown_results + f'---\n'
    
with open(MARKDOWN_FILE, 'w') as f:
    f.write(markdown_results)    
```

```python
MARKDOWN_FILE = './alkitab_data_v2.md'
PDF_FILE = './alkitab_data_v3.pdf'

with open(MARKDOWN_FILE, 'r') as f:
    html_text = markdown.markdown(f.read(), output_format='html5')
    
pdfkit.from_string(html_text, PDF_FILE)
```
